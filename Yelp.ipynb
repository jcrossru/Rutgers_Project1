```python
# Dependencies
import pandas as pd
import os
import numpy as np
import requests
import json
from pprint import pprint
from config import yelp_key
from statistics import mean 
```

```python
#Creating a sample list and dataset
counties = ["Burlington, NJ", "Mercer, NJ"]

summary_df = pd.DataFrame()
summary_df["Counties"] = counties
summary_df["Avg_Rating"] = ""
summary_df["Avg_Price"]=""
# summary_df.rename(columns = "0":"Counties", inplace=True)
summary_df.head()
```

```python
headers = { 'Authorization': 'Bearer {}'.format(yelp_key),}
base_url = "https://api.yelp.com/v3/businesses/search"

response = requests.get(base_url, params = {"location":"Mercer, NJ", "Catergories":"Restaurants", "limit": 5}, headers = headers)
json_data = response.json()
pprint (json_data)
```

```python

avg_rating = []
avg_price = []

# For each county run through the restaurants
for county in counties:
    response = requests.get(base_url, params = {"location":county, "Catergories":"Restaurants", "limit": 5}, headers = headers)
    json_data = response.json()
    #empty lists for data
    ratings = []
    price = []
    #Run through the rows of data and print average
    for row in range (0, 5):
        try:
            ratings.append(json_data["businesses"][row]["rating"])
        except KeyError:
            print("No rating")
        try:
            price.append(len(json_data["businesses"][row]["price"]))
        except KeyError:
            print ("No price")
    avg_rating.append (mean(ratings))
    avg_price.append (mean(price))
#     summary_df["Avg_Rating"][county]= (mean(ratings))
#     summary_df["Avg_Price"][county]= (mean(price))
# summary_df.head()
print (avg_rating)
print (avg_price)
```
