```python
# Dependencies
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import os
import numpy as np
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
#Bring in the csv file created from the Crime Data and convert to a dataframe
csvpath = os.path.join("Yelp_reformat.csv")
#crime_df = pd.read_csv(csvpath, encoding="UTF-8")
yelptest_df = pd.read_csv(csvpath, encoding="UTF-8")
yelptest_df.columns=['County', 'Crime Rate per 100k', 'Murder', 'Rape', ' Robbery', 'Assault', 'Burglary',
                   'Larceny', 'MV Theft', 'Arson', 'Population', 'Avg_Rating','Median_Rating','Avg_Price',
                    'Rest_Count', 'Avg_Lat', 'Avg_Lng', 'CountyName', 'State']
yelptest_df.head()
```

```python
#Compare Average Rating to Murder
yelptest_df.plot.scatter(x='Avg_Rating', y='Crime Rate per 100k')
yelptest_df['Avg_Rating'].corr(yelptest_df['Crime Rate per 100k'])
```

```python
#Compare Average Rating to Murder
yelptest_df.plot.scatter(x='Avg_Rating', y='Murder')
yelptest_df['Avg_Rating'].corr(yelptest_df['Murder'])
```

```python
#Compare Average Rating to Assault
yelptest_df.plot.scatter(x='Avg_Rating', y='Assault')
yelptest_df['Avg_Rating'].corr(yelptest_df['Assault']) 
```

```python
#Compare Avg Rating to Crime Rate
yelptest_df.plot.scatter(x='Avg_Rating', y='MV Theft')
yelptest_df['Avg_Rating'].corr(yelptest_df['MV Theft']) 
```

```python
#Compare Population to Murder
yelptest_df.plot.scatter(x='Population', y='Murder')
yelptest_df['Population'].corr(yelptest_df['Murder'])
```

```python
#Compare Types of Crime
labels = 'Murder', 'Robbery', 'Burglary', 'Larceny', 'Rape', 'Assault', 'MV Theft', 'Arson'
sizes = [10, 10, 10, 10, 10, 10, 10, 10]
explode = (0, 0, 0, 0, 0, 0, 0, 0) 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()

```

```python
#Compare Restaurant Count to Murder
yelptest_df.plot.scatter(x='Rest_Count', y='Murder')
```

```python
#Create a heatmap by state

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# sphinx_gallery_thumbnail_number = 2

vegetables = ["cucumber", "tomato", "lettuce", "asparagus",
              "potato", "wheat", "barley"]
farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening",
           "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])


fig, ax = plt.subplots()
im = ax.imshow(harvest)

# We want to show all ticks...
ax.set_xticks(np.arange(len(farmers)))
ax.set_yticks(np.arange(len(vegetables)))
# ... and label them with the respective list entries
ax.set_xticklabels(farmers)
ax.set_yticklabels(vegetables)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(vegetables)):
    for j in range(len(farmers)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Avg Yelp Rating by State")
fig.tight_layout()
plt.show()
```
