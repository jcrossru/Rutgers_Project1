```python
# Dependencies
import pandas as pd
import os
import numpy as np
import requests
import json
from pprint import pprint
from config import yelp_key
from statistics import mean, median 
```

```python
#Bring in the csv file created from the Crime Data and convert to a dataframe
csvpath = os.path.join("Yelp_data.csv")
crime_df = pd.read_csv(csvpath, encoding="UTF-8")
crime_df.head()
```

```python
#add new columns to dataframe
crime_df["CountyName"] = crime_df.County.str.slice(0, -4)
crime_df["State"] = crime_df.County.str.slice(-2)
crime_df.head()
```

```python
# Convert to csv
crime_df.to_csv("Yelp_reformat.csv", index=False, header=True)
```
